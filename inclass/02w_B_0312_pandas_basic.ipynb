```python
import numpy as np
import pandas as pd

s= ['희빈','수영', '현수', '지호','지연']
p = pd.Series([99,84,92,65,78], index=s)
p
```

```python
p.name = '수학성적'
p.index.name = 'st_name'
p
```

```python
p.sort_values()
```

```python
p.sort_values(ascending=False)
```

```python
p.sort_index()
```

```python
p.sort_index(ascending=False)
```
