```python
df3 = pd.DataFrame({'A': range(1,6) , 'B' : range(6,-3 , -2),'C': range(10,5,-1),'D clmns':range(3,-2,-1), 'E':['python','java','python','go','python']})
df3
```

```python
df3.query('A > B')
```

```python
df3[df3.A > df3.B]
```

```python
df3.query('C==10')
```

```python
df3.query('A>0')
```

```python
df3.query('E =="python"')
```

```python
df3.query('E !="python"')
```

```python
df3.query('A ==D clmns')
```

```python
df3.query('A =="D clmns"')
```

```python
df3.query("A =='D clmns'")
```

```python
df3[df3.A == df3['D clmns']]
```

```python
df3.query('A > 3 | B > 8')
```

```python
df3.query('A > 2 | C > 8')
```

```python
df3.query('E in ["python", "go"]')

```

```python
num = 3
df3.query('B > @num')
```

```python
num = 3
df3.query(f'B > {num}')
```

```python
s = "E.str.contains('py')"
df3.query(s)
```

```python
s = "E.str.startswith('g')"
df3.query(s)
```
