```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/news_comment_BTS.csv" , encoding="utf-8")
df
```

```python
df.head()
```

```python
df['reply'][:10]
```

```python
df['reply'] = df['reply'].str.replace(r'[^가-힣]', ' ', regex=True)
df['reply'].head()
```

```python
from konlpy.tag import Kkma
kkma = Kkma()
kkma.nouns("질문이나 건의사항은 깃헙 이슈 트래버에 남겨 주세요.")
```

```python
df2 = pd.DataFrame({"A":[1,2,3] , "B":[4,5,6]})
df2
```

```python
df_s = df2.apply(lambda x: x**2)
df_s
```

```python
form = lambda x: x**2
df_s = df2.apply(form)
df_s
```

```python
nonus = df['reply'].apply(kkma.nouns)
nonus
```

```python
df3 = pd.DataFrame({"A": [[1,2,3] , "foo" ,[],[3,4] ] , "B": 1 , "C": [["A","B","C"] , np.nan ,[] , ["d" , "e"]]})
df3
```

```python
df3.explode("A")
```

```python
df3.explode(["A" ,"C"])
```

```python
type(nonus)
```

```python
nonus = nonus.explode()
nonus
```

```python
nonus
```

```python
df_word1 = pd.DataFrame({"word": nonus})
df_word1.head()
```

```python
df_word1['len'] = df_word1['word'].str.len()
df_word1 = df_word1.query('len >= 2')
df_word1.head()
```

```python
df_freq = df_word1.groupby("word" , as_index=False).\
            agg(n=("word", "count")). \
            sort_values(by="n", ascending=False)
df_freq
```

```python
top20 = df_freq.head(20)
top20
```

```python
sns.barplot(data=top20, y="word", x="n" , hue="word")
```

```python
from PIL import Image
icon = Image.open('data/cloud.png')

img = Image.new('RGB', icon.size, (255, 255, 255))
img.paste(icon, icon)
img = np.array(img)
img;
```

```python
dic_word = df_freq.set_index("word").to_dict()["n"]
dic_word
```

```python
from wordcloud import WordCloud
wc = WordCloud(font_path=FONT_PATH, background_color='white',
               random_state = 1234, width=400, height=400, colormap='plasma',
               mask = img)
```

```python
img_wc = wc.generate_from_frequencies(dic_word)
plt.figure(figsize=(6, 4))
plt.axis('off')
plt.imshow(img_wc);
```
